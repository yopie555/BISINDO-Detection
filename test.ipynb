```python
import cv2
from ultralytics import YOLO

# Load pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')  # 'n' is the nano model, you can use 's', 'm', etc.
```

```python
# Initialize video capture for webcam
cap = cv2.VideoCapture(0)  # 0 is the default webcam
```

```python
while True:
    ret, frame = cap.read()  # Capture frame-by-frame
    
    if not ret:
        break  # If there's an issue with the camera, exit the loop
    
    # Use the model to make predictions on the current frame
    results = model(frame)  # YOLOv8 inference
    
    # Visualize the results
    annotated_frame = results[0].plot()  # Annotates the frame with bounding boxes, labels, etc.
    
    # Display the frame
    cv2.imshow('YOLOv8 Real-Time Object Detection', annotated_frame)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
